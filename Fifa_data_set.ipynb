```python
db1 = db[["Name", "Age", "Overall", "Position", "Wage", "Value", "Club", "Nationality"]]
```

```python
db1.head()
```

```python
def price(abc):
    if abc[-1].isdigit():
        return float(abc[1:])
    else:
        res = float(abc[1:-1])
        if abc[-1] == 'K':
            return res*1000
        elif abc[-1]=='M':
            return res*1000000
```

```python
db1["Wage"] = db1["Wage"].apply(price)
db1["Value"] = db1["Value"].apply(price)
```
